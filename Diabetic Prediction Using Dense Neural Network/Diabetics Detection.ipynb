# PIMA DIABETES INDIAN
PROBLEM DESCRIPTION
Main aim of this project is to detect wether the women who is pregnant is Diabetic or not 

So basically we have collected data from open Source So that we can detect the possibility of Diabetic
Here the Data Contains 9 Column
Description of dataset
   1. Number of times pregnant
   2. Plasma glucose concentration a 2 hours in an oral glucose tolerance test
   3. Diastolic blood pressure (mm Hg)
   4. Triceps skin fold thickness (mm)
   5. 2-Hour serum insulin (mu U/ml)
   6. Body mass index (weight in kg/(height in m)^2)
   7. Diabetes pedigree function
   8. Age (years)
   9. Class variable (0 or 1)

```python
# importing basic libaries for the automation of data and reading the data 
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
from keras.models import model_from_json
```

```python
#Reading the data from the csv file
dataset = loadtxt('pima-indians-diabetes.csv', delimiter=',')
```

```python
x = dataset[:,0:8]
y = dataset[:,8]
```

```python
#neuron building 
model = Sequential()
model.add(Dense(12,input_dim = 8 , activation='relu'))#Hidden Layer
model.add(Dense(8,activation = 'relu'))#hidden layer
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
#Optmizing the data to get efficent and accurate neural network
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x, y, epochs=300, batch_size=220)
```

```python
#accuracy
_, accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
#saving the model on our current directiory
model_json = model.to_json()
with open("model.json","w") as json_file:
    json_file.write(model_json)
model.save_weights('model.h5')
print("model Saved to disk")
```

```python
json_model = open('model.json','r')
loaded_model = json_model.read()
json_model.close()
model = model_from_json(loaded_model)
model.load_weights("model.h5")
print("Loaded model from disk")
```

```python
#testing the model to get accuracy 
predictions = model.predict_classes(x)
for i in range(5,20):
    print('%s => %d (expected %d)' % (x[i].tolist(), predictions[i], y[i]))
```
